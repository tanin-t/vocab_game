```python
# todo
import csv

with open('data/vocab_data.csv') as f:
    rows = csv.DictReader(f)
    for row in rows:

        vocab = Vocab()
        # for key in row:
        #     setattr(vocab, key, row['key'])
        # vocab.vocab_set_id = int(row['vocab_set_id']) + 7
        # vocab.save()

        vocab.vocab = row['vocab']
        vocab.vocab_set_id = int(row['vocab_set_id']) + 7
        vocab.part_of_speech = row['part_of_speech']
        vocab.meaning = row['meaning']
        vocab.picture = row['picture']
        vocab.sound = row['sound']
        vocab.phonetic = row['phonetic']
        vocab.save()
```
